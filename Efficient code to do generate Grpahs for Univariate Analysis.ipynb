```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris.columns
```

```python
newiris = pd.melt(iris, id_vars = ['species'])
newiris.head()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(newiris, col = 'variable', hue = 'species',palette="Set1")\
   .map(sns.distplot,"value", hist=False)\
   .add_legend();
```
